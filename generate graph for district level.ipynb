```python
def generateGraph(index):
    f = plt.figure(figsize=(11.69,8.27))
    #f.suptitle('This is the main title', fontsize=18) 
    
    # set width of bar
    barWidth = 0.25

    # for first graphset height of bar
    bars1 = [e1[index], w1[index], f1[index], l1[index], m1[index], s1[index], d1[index]]
    bars2 = [e2[index], w2[index], f2[index], l2[index], m2[index], s2[index], d2[index]]
    bars3 = [e3[index], w3[index], f3[index], l3[index], m3[index], s3[index], d3[index]]

    # for second graphset height of bar
    bars4 = [e4[index], w4[index], f4[index], l4[index], m4[index], s4[index], d4[index]]
    bars5 = [e5[index], w5[index], f5[index], l5[index], m5[index], s5[index], d5[index]]
    bars6 = [e6[index], w6[index], f6[index], l6[index], m6[index], s6[index], d6[index]]

    # for third graphset height of bar
    bars7 = [e7[index], w7[index], f7[index], l7[index], m7[index], s7[index], d7[index]]
    bars8 = [e8[index], w8[index], f8[index], l8[index], m8[index], s8[index], d8[index]]
    bars9 = [e9[index], w9[index], f9[index], l9[index], m9[index], s9[index], d9[index]]

    # for third graphset height of bar
    bars10 = [e10[index], w10[index], f10[index], l10[index], m10[index], s10[index], d10[index]]
    bars11 = [e11[index], w11[index], f11[index], l11[index], m11[index], s11[index], d11[index]]
    bars12 = [e12[index], w12[index], f12[index], l12[index], m12[index], s12[index], d12[index]]

    # Set position of bar on X axis
    r1 = np.arange(len(bars1))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]

    # Make the plot
    plt.subplot(221)
    plt.title('Buiding exposure(number)')
    plt.bar(r1, bars1, color='#00b04f', width=barWidth, edgecolor='white', label='low')
    plt.bar(r2, bars2, color='#ffbf00', width=barWidth, edgecolor='white', label='medium')
    plt.bar(r3, bars3, color='#ff0000', width=barWidth, edgecolor='white', label='high')

    # Add xticks on the middle of the group bars
    # plt.xlabel('group', fontweight='bold')
    plt.xticks([r + barWidth for r in range(len(bars1))], ['EQ', 'WS', 'FL', 'LS', 'MF', 'SA', 'DR'])


        # Make the plot
    plt.subplot(222)
    plt.title('Population(number)')
    plt.bar(r1, bars4, color='#00b04f', width=barWidth, edgecolor='white', label='low')
    plt.bar(r2, bars5, color='#ffbf00', width=barWidth, edgecolor='white', label='medium')
    plt.bar(r3, bars6, color='#ff0000', width=barWidth, edgecolor='white', label='high')

    # Add xticks on the middle of the group bars
    # plt.xlabel('group', fontweight='bold')
    plt.xticks([r + barWidth for r in range(len(bars1))], ['EQ', 'WS', 'FL', 'LS', 'MF', 'SA', 'DR'])

    # Make the plot
    plt.subplot(223)
    plt.title('Agricultural land exposed(hectares)')
    plt.bar(r1, bars7, color='#00b04f', width=barWidth, edgecolor='white', label='low')
    plt.bar(r2, bars8, color='#ffbf00', width=barWidth, edgecolor='white', label='medium')
    plt.bar(r3, bars9, color='#ff0000', width=barWidth, edgecolor='white', label='high')

    # Add xticks on the middle of the group bars
    # plt.xlabel('group', fontweight='bold')
    plt.xticks([r + barWidth for r in range(len(bars1))], ['EQ', 'WS', 'FL', 'LS', 'MF', 'SA', 'DR'])


    # Make the plot
    plt.subplot(224)
    plt.title('Road exposure(km)')
    plt.bar(r1, bars10, color='#00b04f', width=barWidth, edgecolor='white', label='low')
    plt.bar(r2, bars11, color='#ffbf00', width=barWidth, edgecolor='white', label='medium')
    plt.bar(r3, bars12, color='#ff0000', width=barWidth, edgecolor='white', label='high')

    # Add xticks on the middle of the group bars
    # plt.xlabel('group', fontweight='bold')
    plt.xticks([r + barWidth for r in range(len(bars1))], ['EQ', 'WS', 'FL', 'LS', 'MF', 'SA', 'DR'])


    # Create legend & Show graphic
    plt.tight_layout()
    plt.savefig('output/figure-{}'.format(index), figsize=(11.69,8.27))
```
